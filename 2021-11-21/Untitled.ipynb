```python
import pandas as pd
import numpy as py
import sys, os
```

```python
input_path='/fastscratch/c-panz/2021-11-21'
df_lambda_5mC = pd.read_csv(os.path.join(input_path, 'lambda_5mC.test.tsv'), header=None, sep ='\t')
df_lambda_5C = pd.read_csv(os.path.join(input_path, 'lambda_5C.test.tsv'), header=None, sep ='\t')
df_lambda_5mC
```

# Split train and test

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold, GridSearchCV, RepeatedStratifiedKFold
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import EditedNearestNeighbours, RandomUnderSampler
from imblearn.pipeline import Pipeline as imbpipeline
from imblearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, f1_score

from collections import Counter


df_lambda_5mC_train, df_lambda_5mC_test = train_test_split(df_lambda_5mC,
                                                           train_size=0.5,
                                                           random_state=42)

df_lambda_5C_train, df_lambda_5C_test = train_test_split(df_lambda_5C,
                                                         train_size=0.5,
                                                         random_state=42)
```

```python
df_lambda_5mC_train.shape, df_lambda_5C_train.shape
```

```python
X_train = pd.concat([df_lambda_5mC_train, df_lambda_5C_train], axis=0).reset_index(drop=True)
X_test = pd.concat([df_lambda_5mC_test, df_lambda_5C_test], axis=0).reset_index(drop=True)
```
